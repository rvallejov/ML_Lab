```python
from __future__ import print_function  # Python 2 and 3
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import re
from scipy import stats
%matplotlib inline
plt.style.use('ggplot')
from datetime import datetime
```

```python
 orders = pd.read_csv('data/Orders.csv', index_col=0)
```

```python
orders.hist(bins=50, figsize=(20,15))
```

```python
orders.columns
```

```python
orders['Sales'] = orders['Sales'].replace('[\$,]', '', regex=True).astype(float)
orders['Profit'] = orders['Profit'].replace('[\$,]', '', regex=True).astype(float)
```

```python
orders.head()
```

```python
orders['Order.Date'] = pd.to_datetime(orders['Order.Date'])
```

```python
orders_group = orders.groupby('Order.Date').agg({'Quantity':sum})
```

```python
orders_group.head()
```

```python
orders['Year'] = list(map(lambda x: x.year, orders['Order.Date']))
orders['Month'] = list(map(lambda x: x.month, orders['Order.Date']))
orders['Day'] = list(map(lambda x: x.day, orders['Order.Date']))
```

```python
orders.head()
```

```python
orders_group_month = orders.groupby('Month').agg({'Quantity':'mean'})
```

```python
orders_group_month = orders_group.reset_index()
```

```python
orders_group_month.columns
```

```python
plt.plot(orders_group_month['Month'], orders_group_month['Quantity'])
```

```python
orders.columns
```

```python
orders_group_category = orders.groupby(['Year', 'Month', 'Category']).agg({'Quantity':'sum'})
orders_group_category = orders.groupby(['Month', 'Category']).agg({'Quantity':'mean'})
orders_group_category = orders_group_category.reset_index()
```

```python
for i in set(orders_group_category['Category']):
    plt.plot(orders_group_category[orders_group_category['Category'] == i]['Month'], 
             orders_group_category[orders_group_category['Category'] == i]['Quantity'])
    plt.show()
```
